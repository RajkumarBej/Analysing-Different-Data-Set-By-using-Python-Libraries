Distribution plot:there are tow type of variable, ie;1.quantitative variable which can't be categorise and has multiple values. and 2.qualitative variable can be categorise and limited value but repeated in nature.

Distribution plots are based on quantitative data or variable
eg;1.distplot/histplot
2.jointplot
3.pairplot
4.rugplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('tips')
df.head()
```

in the above dataset sex,smoker,day,time and size are qualtative variable or categorical variable and we can't apply distibution plot directly.

```python
sns.jointplot(x='total_bill',y='tip',data=df)#kind=scatter,hex,reg
```

```python
sns.pairplot(df,hue='size',palette='rainbow')
```

```python
sns.rugplot(df)
```

Categorical data plots:boxplot,violinplot,stripplot,swarmplot,barplot,countplot

bar plot: in barplot we provid quantitative variable in y-axis and qualitative variable in x-axis

```python
sns.barplot(x='size',y='tip',data=df)
```

```python
sns.boxplot(x='sex',y='tip',data=df)
```

Matrix plot:heat map,cluster map and pivot table heat map

heat map:it is created by interpreting corelation(corelation is find from covarience and coverience lies from -1 to 1)

these all matrix plots are deal with numeric value

```python
nvar=df[['total_bill','tip','size']]
nvar
```

```python
nvar.corr()
```

```python
sns.heatmap(nvar.corr(),annot=True)
```

```python
sns.clustermap(nvar.corr())
```
