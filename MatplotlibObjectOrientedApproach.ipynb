```python
import matplotlib.pyplot as plt
```

```python
data={
    'id':[101,105,106,109,112,116,128,158],
    'name':['rabi','mithun','pinku','raju','dilip','dipak','surya','pintu'],
    'age':[21,25,22,21,23,26,28,23],
    'dept':['hr','it','finance','hr','it','hr','finance','it'],
    'salary':[1000,3200,4000,5400,1200,1500,4000,2300]
}
print(data)
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(data)
df
```

```python
#plt.subplot(nrow,ncol,idx)
plt.subplot(1,2,1)
plt.plot(df['id'],df['salary'],label='id and salary')
plt.xlabel("id")
plt.ylabel('salary')
plt.title("Line plot")
plt.legend()

plt.subplot(1,2,2)
df1=df['salary'].groupby(df['dept'])
df1
plt.bar(df['name'],df['salary'],color='red',label='name and salary')
plt.xlabel('name')
plt.ylabel('salary')
plt.title('bar plot')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#fig, axi=plt.subplots(nrow,ncol,fig=(width,height))
fig, axi=plt.subplots(2,3,figsize=(10,5))
axi[0][1].plot(df['id'],df['salary'])
axi[0][1].set_title('line plot')

axi[0][2].hist(df['age'],bins=5)
axi[0][2].set_title('histogram')

dept_cout=df['dept'].value_counts()
axi[1][0].pie(dept_cout)
axi[1][0].set_title('pie chart')

fig.suptitle("charts")
plt.tight_layout()
plt.show()
```
