All function that are used in matplotlib can be used with seaborn

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import numpy as np
```

```python
sns.get_dataset_names()
```

```python
df_data=sns.load_dataset('penguins')
```

```python
print(df_data.head())
```

```python
print(df_data.dtypes)
print(df_data.info)
print(df_data.describe)
```

```python
sns.lineplot(data=df_data,x='species',y='body_mass_g')
```

```python
sns.lineplot(x=df_data['species'],y=df_data['body_mass_g'])
```

```python
sns.lineplot(data=df_data,x='bill_length_mm',y='bill_depth_mm')
```

hue= (This function plot graph with respect to popular data in dataframe)

```python
sns.lineplot(data=df_data,x='bill_length_mm',y='bill_depth_mm',hue='sex')
```

size= (it is used to adjust the size of plots)

style= (it is used to change style of plot)

```python
sns.lineplot(data=df_data,x='bill_length_mm',y='bill_depth_mm',hue='sex',size=100,style='sex')
```

palette='color name' (it is used to set color of plots. it has different color list to check them you have to generate error by passing wrong value.)

markers='' (it is used to set a symbol at plot deviation point )

```python
sns.lineplot(data=df_data,x='bill_length_mm',y='bill_depth_mm',hue='sex',palette='Accent',markers=["o",">"])
```

```python
sns.lineplot(data=df_data,x='bill_length_mm',y='bill_depth_mm',hue='sex',style='sex')
```

dashes=True/False (It is used to remove dashes)

legend='true/false/auto/full'

```python
sns.lineplot(data=df_data,x='bill_length_mm',y='bill_depth_mm',hue='sex',style='sex',dashes=False,legend='full')
plt.grid()
plt.title("bill_length_mm vs bill_depth_mm")
plt.show()
```

```python
sns.barplot(data=df_data,x='species',y='bill_length_mm',hue='sex')
```

order='list name' (it is used to set the order)

```python
order_list=['Chinstrap','Gentoo','Adelie']
sns.barplot(data=df_data,x='species',y='bill_length_mm',hue='sex',order=order_list,hue_order=['Female','Male'])
plt.show()
```

```python
order_list=['Chinstrap','Gentoo','Adelie']
sns.barplot(data=df_data,x='species',y='bill_length_mm',hue='sex',order=order_list,hue_order=['Female','Male'],
            ci=100,n_boot=6)
plt.show()
```

orient='value' (to do horizontal bar)

```python
sns.barplot(data=df_data,x='bill_length_mm',y='bill_depth_mm',palette='Accent',orient='h')
```

saturation='value' (to change font light,dark etc.)

```python
sns.barplot(data=df_data,x='species',y='bill_length_mm',hue='sex',saturation=5,errcolor='red',errwidth=10,
            capsize=0)
plt.show()
```

dodge=True/False (to do overlapping of bars)

```python
sns.set(style='darkgrid')
sns.barplot(data=df_data,x='species',y='bill_length_mm',hue='sex',saturation=100,alpha=0.4)
plt.show()
```

Histogram: sns.distplot(df[''])

```python
sns.displot(df_data["bill_length_mm"])
plt.show()
```

bin= (it is used to align the bar within particular range)

```python
sns.displot(df_data["bill_length_mm"],bins=[35,40,45,50,55,60],color='Red')
plt.show()
```

```python
sns.displot(df_data['bill_depth_mm'],kde=True,rug=True)
plt.show()
```

log_scale=True/False (to convert the data into log)

```python
sns.displot(df_data['bill_depth_mm'],kde=True,rug=True,log_scale=True)
plt.show()
```

scatter plot= sns.scatterplot(x='',y='',data=df_name)

```python
sns.scatterplot(data=df_data,x='bill_length_mm',y='bill_depth_mm')
plt.show()
```

```python
sns.scatterplot(data=df_data,x='bill_length_mm',y='bill_depth_mm',hue='sex',style='sex',size='sex',sizes=(100,20),palette='OrRd_r')
plt.show()
```

```python
mark={"Male":"*","Female":"o"}
sns.scatterplot(data=df_data,x='bill_length_mm',y='bill_depth_mm',hue='sex',palette='OrRd_r',markers=mark,alpha=0.5)
plt.show()
```
