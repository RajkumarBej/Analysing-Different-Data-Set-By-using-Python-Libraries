```python
import pandas as pd
```

```python
import numpy as np
```

```python
data={
    "name":['alice','bob','charie','david','eva','alice'],
    "age":[20,30,35,np.nan,29,25],
    "department":['Hr','IT','Finance','IT','HR','HR'],
    "salary":[50000,60000,70000,62000,np.nan,50000]
}
print(data)
```

```python
df=pd.DataFrame(data)
print(df)
```

adding a new column:insert()

```python
#df.insert(loc,column name,data list)
df.insert(0,'id',[101,104,321,122,231,452])
df
```

updating value:loc()

```python
#df.loc[row index,'column name']=value
df.loc[3,'salary']=75000
df
```

null value checking

```python
print(df.isnull())
print(df.isnull().sum())
```

drop null value:

```python
#df.dropna(axis=0,inplace=True) for rows
#df.dropna(axis=1,inplace=True) for columns
df.dropna(axis=0)
```

```python
df
```

filling null value:-fillna

```python
#df.fillna(value,inplace=True) you can give mean,median in place of value
df.fillna(0)
```

```python
#df['age'].fillna(df['age'].mean(),inplace=True)
df['age'].fillna(df['age'].mean())
```

Interpulation:filling na with estimate value(preserve data integrity,smooth trands,avoid data loss):- interpolate()

```python
#df.interpolate(method="linear or polynomial or time or quadratic or qubic or etc. ",axis=0,inplace=True)
df.interpolate(method='linear',axis=0)
#uesed with:-
#Timer series data
#numeric data with trends
#avoid dropping rows
```

```python
df['age'].interpolate(method='cubic',axis=0)
```

```python
df.sort_values(by='age',ascending=True)
```

Grouping the data:

```python
#group_obj_name=df.groupby("col1")["col2"].aggregating function
group_obj=df.groupby('department')['salary'].sum()
print(group_obj)
```

Merging:

```python
#pd.merge(dataframe1,dataframe2,on='column name',how='type of join')
data2={
    "name":['charlie','mark','bob','david','lorence'],
    "donation":[500,300,1000,4000,600]
}
df2=pd.DataFrame(data2)
print(data)
print(df2)
```

```python
#pd.merge(dataframe1,dataframe2,on='column name',how='type of join')
df_merge=pd.merge(df,df2,on='name',how='inner')
print(df_merge)
df_merge=pd.merge(df,df2,on='name',how='outer')
print(df_merge)
df_merge=pd.merge(df,df2,on='name',how='left')
print(df_merge)
df_merge=pd.merge(df,df2,on='name',how='right')
print(df_merge)
```

```python
df_merge=pd.merge(df,df2,how='cross')
df_merge
```

Data frame concatination

```python
#concating the data frames vertically(row wise) and horizontally(column wise)
#pd.concate([dataframe1,dataframe2],axis=0,ignore_index=True)
df_concat=pd.concat([df,df2],axis=0,ignore_index=True)#vertically
print(df_concat)
df_concat=pd.concat([df,df2],axis=1,ignore_index=False)#horizontally
print(df_concat)
```
