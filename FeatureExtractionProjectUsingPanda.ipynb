```python
anime=pd.read_csv(r'anime.csv')
anime.head(5)
```

create a new column for episod

```python
def extract_episod(text):
    data=""
    flag=False
    for i in text:
        if i==')':
            break
        if i== '(':
            flag=True
            continue
        if flag:
            data+=i
    return data
```

```python
anime['episods']=anime['Title'].apply(extract_episod)
anime.head()
```

```python
anime['episods']=anime['episods'].str.replace("eps","")
anime.head()
```

Make a new column for time stamp

```python
def extraction_of_time(text):
    flag=False
    data=""
    count=0
    for i in text:
        if count==19:
            break
        if i==')':
            flag=True
            continue
        if flag:
            data=data+i
            count+=1
    return data
```

```python
temp=extraction_of_time(anime.loc[0]['Title'])
temp
```

```python
anime['time stamp']=anime['Title'].apply(extraction_of_time)
anime.head()
```
