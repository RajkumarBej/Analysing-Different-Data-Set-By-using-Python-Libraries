```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
netflix_data=pd.read_csv('netflix_titles.csv')
netflix_data.head(5)
```

```python
netflix_data.dropna(subset=['type','director','cast','country','release_year','rating','duration'])
```

```python
type_count=netflix_data['type'].value_counts()
```

```python
type_count
```

```python
plt.figure(figsize=(6,4))
plt.bar(type_count.index,type_count.values,color=['red','blue'])
plt.xlabel('Type')
plt.ylabel('Number')
plt.title('Movie vs TV Show')

plt.tight_layout()
plt.savefig('C:/Users/surya/Desktop/CODES/MachineLearning/NetflixImg1.png')
plt.show()
```

```python
rating_count=netflix_data['rating'].value_counts()
plt.figure(figsize=(8,6))
plt.pie(rating_count,labels=rating_count.values,autopct='%1.11f%%')
plt.title("Rating")

plt.tight_layout()
plt.show()
```

```python
netflix_data.head()
```

```python
movie_df=netflix_data[netflix_data['type']=='Movie'].copy()
movie_df.dropna(subset=['type','director','cast','country','release_year','rating','duration'],inplace=True)
movie_df['duration_int']=movie_df['duration'].str.replace(' min','').astype(int)

plt.figure(figsize=(6,4))
plt.hist(movie_df['duration_int'],bins=50,color='orange')
plt.xlabel('number')
plt.ylabel('duration')
plt.title('netflix duration data')
plt.tight_layout()
plt.savefig('C:/Users/surya/Desktop/CODES/MachineLearning/NetflixImg2.png')
plt.show()
```

```python
release_count=netflix_data['release_year'].value_counts().sort_index()

plt.figure(figsize=(6,4))
plt.scatter(release_count.index,release_count.values,color='green')
plt.xlabel('release year')
plt.ylabel('release count')
plt.title('Release count of movies and tv shows with respect to year')

plt.tight_layout()
plt.show()
```

```python
country_list=netflix_data['country'].dropna().value_counts().head(10)
country_list

plt.figure(figsize=(10,4))
plt.bar(country_list.index,country_list.values)
plt.xlabel('name')
plt.ylabel('index')
plt.title('top 10 country vs shows')
plt.tight_layout()
plt.show()
```

```python
content_by_year=netflix_data.groupby(['release_year','type']).size().unstack().fillna(0)
plt.figure(figsize=(6,4))

fig, ax=plt.subplots(1,2,figsize=(6,4))
ax[0].plot(content_by_year.index,content_by_year['Movie'],color='red')
ax[0].set_title('year of release vs Movies and TV Shows')

ax[0].plot(content_by_year.index,content_by_year['TV Show'],color='blue')
#ax[0].set_title('year of release vs TV Shows')
plt.tight_layout()
plt.show()
```
