Pair plot: sns.pairplot(data frame name) 

```python
var=sns.load_dataset("tips")
var.head(5)
```

```python
sns.pairplot(var)
```

```python
sns.pairplot(var,hue="sex")
```

```python
sns.pairplot(var,vars=["total_bill","tip"],hue="sex",hue_order=['Female','Male'],palette="BuGn")
```

```python
sns.pairplot(var,x_vars=['total_bill','tip'],hue="sex")
```

```python
sns.pairplot(var.head(10),y_vars=['total_bill','tip'],hue="sex")
```

kind='' (it is used to define graph structure or plotation) eg:-'scatter','reg','ked','hist'

```python
sns.pairplot(var,x_vars=['total_bill','tip'],hue="sex",kind="reg")
```

```python
sns.pairplot(var.head(20),hue='sex',kind="kde")
```

```python
sns.pairplot(var.head(20),diag_kind='kde',hue="sex",kind='scatter',markers=['*','<'])
```

Strip Plot: it is basically a scatter plot that differentiates different categories. Strip plots are considered a good alternative to a box plot or a violin plot.

sns.stripplot(x='',y='',data='')

```python
sns.stripplot(data=var,y='total_bill',x='day',hue='sex',palette='rocket_r')
```

```python
sns.stripplot(data=var,y='total_bill',x='day',hue='sex',linewidth=1,edgecolor='b',jitter='5')
```

```python
sns.stripplot(data=var,y='total_bill',x='day',hue='sex',size=5,marker='*')
```

Box plot: sns.boxplot(data='',x='',y='')

```python
var.head(10)
```

```python
sns.boxplot(data=var,y='total_bill',x='day',hue='sex',color='g',order=['Sat','Fri','Thur','Sun'])
plt.show()
```

```python
sns.boxplot(data=var,x='day',y='total_bill',showmeans=True,meanprops={'marker':'+','markeredgecolor':'r'},width=0.5,color='w')
```

Cat plot: sns.catplot(data='',x='',y='',kind='') using cat plot multiple graph can be generated using kind=''.

categorical scatterplots with catplot:1.strip()-with kind="strip" 2.swarmplot()-with kind="swarm"

Categorical distribution plots with catplot:1.kind="box" 2.kind="violin" 3.kind="boxen"(for boxenplot())

Categorical estimate plots with catplot: 1.kind="poind" 2.kind="bar" 3.kind="count"

```python
sns.catplot(data=var.head(10),x="tip",y="size",kind="swarm",hue="sex")
```

Styling plots:Seaborn Figure Styles,removing axes spines and scale and context

```python
sns.set_style("dark")
sns.barplot(data=var,x='day',y='tip')
plt.show()
```

```python
sns.set_style("darkgrid")#to change background style
sns.barplot(data=var,x='day',y='tip')
plt.show()
```

```python
sns.set_style("whitegrid")
sns.barplot(data=var,x='day',y='tip')
sns.despine()#to remove axes line
plt.show()
```

```python
sns.set_style("dark")
plt.figure(figsize=(6,4))#set size or scalling
sns.barplot(data=var,x='day',y='tip')
plt.show()
```

```python
sns.set_style("dark")
#sns.set_context("type(poster/paper)",font_scale=value,color/pallete)
sns.set_context("paper",font_scale=2)
sns.barplot(data=var,x='day',y='tip')
plt.show()
```
