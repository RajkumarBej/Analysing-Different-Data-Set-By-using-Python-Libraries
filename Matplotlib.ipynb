```python
data={"salary":[25000,34000,67800,16900,34050,16000,62000]}
print(data)
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(data)
df
```

Plot the data frame in graph:

```python
plt.plot(df['salary'],color='red',marker="o",linestyle="--",linewidth=2)
plt.grid()
plt.show()
```

bivarient analysis-numerical and catagorical

```python
hr_salary=df[df['dept']=='HR']['salary']
hr_salary
```

```python
it_salary=df[df['dept']=='IT']['salary']
it_salary
```

```python
manager_salary=df[df['dept']=='Manager']['salary']
manager_salary
```

```python
plt.boxplot([hr_salary,it_salary,manager_salary],tick_labels=['HR','IT','Manager'])
plt.legend()
plt.show()
```

pie chart:

```python
salary_by_dept=df.groupby('dept')['salary'].sum()
salary_by_dept
```

```python
plt.pie(salary_by_dept,labels=salary_by_dept.index,autopct='%1.2f',shadow=True,explode=[0.1,0,0.1,0.2])
plt.show()
```

bar plot:-

```python
hr_mean=sum(hr_salary)/len(hr_salary)
it_mean=sum(it_salary/len(it_salary))
manager_mean=sum(manager_salary/len(manager_salary))
```

```python
plt.bar(['HR','IT','Manager'],[hr_mean,it_mean,manager_mean],color=['red','blue','green'])
plt.grid()
plt.show()
```

Bubble plot:-

```python
df['experience']=[0.1,0.4,0.9,0.5,1.1,1.8,2]
df
```

```python
plt.scatter(df['age'],df['salary'],s=df['experience']*80,color='purple',edgecolors='red')
plt.show()
```

2 newmerical and 1 categorical column

```python
plt.scatter(df['age'],df['salary'],c=df['dept'].map({'HR':'red','Manager':'yellow','IT':'pink','finance':'blue'}))
plt.xlabel('age')
plt.ylabel('salary')
plt.title('Age vs salary vs dept:')
plt.legend()
plt.show()
```

```python
color={'HR':'red','Manager':'yellow','IT':'pink','finance':'blue'}

for dept,color in color.items():
    df_dept=df[df['dept']==dept]
    plt.scatter(df_dept['age'], df_dept['salary'],c=color,label=dept)
plt.legend()
```
