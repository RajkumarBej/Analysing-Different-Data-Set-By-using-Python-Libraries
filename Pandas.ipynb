Dictionary to panda data set

```python
fruit_protien={'avocado':2.0,'guava':2.6,'oranges':0.9,'banana':1.1,'apple':0.3}
s2=pd.Series(fruit_protien)
s2.name='calories'
s2
```

Conditional selection:(and, or and not)

```python
s2[(s2>0.5) & (s2<=2.0)]
```

```python
s2[(s2>0.5) | (s2<=2.0)]
```

```python
s2[~(s2<1.1)]
```

Modifing series data

```python
s2['apple']=0.1
s2[['apple','guava','avocado']]=[0.3,2.5,2.0]
s2
```

b> Fill the missing values with relative value:-

```python
df.fillna(0)
```

We have tow way to use fillna() method. ie;
a>forward fill

```python
df["age"].fillna(method="ffill")
```

b> backward fill

```python
df["age"].fillna(method="bfill")
```
