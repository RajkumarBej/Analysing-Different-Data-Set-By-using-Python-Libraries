```python
import pandas as pd
import numpy as np
```

```python
data={
    "name":['alice','bob','charie','david','eva','alice'],
    "age":[20,30,35,np.nan,29,25],
    "department":['Hr','IT','Finance','IT','HR','HR'],
    "salary":[50000,60000,70000,62000,np.nan,50000]
}
print(data)
```

```python
df=pd.DataFrame(data)
df
```

select specific area data

```python
df.loc[2:4][["age","salary"]]
```

find out name those have above 22 year

```python
df[df["age"]>22]
```

```python
df[(df["age"]>22) & (df["department"]== "HR")]
```

fill particular value in each column missing place 

```python
values={"name":"absent","age":1,"depatment":"null","salary":100}
df.fillna(value=values)
```

Groupby function use

Remember you can't see the values of object created after use of groupby function. to see you have to use agregate funtions

```python
group=df.groupby("department")
for i,v in group:
    print(i)
    print(v)
```

```python
group=df.groupby("department")["salary"].sum()
group
```

```python
group=df.groupby(["department","name"])["salary"].sum()
group
```

Pivot Table: in pivot tables values= always numeric for agrigation

```python
pd.pivot_table(df,values="salary",index="department",columns="name")
```

```python
pd.pivot_table(df,values="salary",index="department",columns="name",aggfunc=sum)
```

```python
pd.pivot_table(df,values=["salary","age"],index="department",columns="name")
```

```python
pd.crosstab(df["department"],df["name"])
```
