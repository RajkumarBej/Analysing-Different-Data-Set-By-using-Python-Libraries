Creating some numpy arrays

Analysing a numpy array

```python
x=np.random.randint(1,100,(5,8))
print(x)
print(x.shape)
print(x.dtype)
print(x.ndim)
print(x.size)
```

Transpose of a matrix

```python
a=np.random.randint(1,1000,(6,4))
print(a)
print(a.shape)
print(a.ndim)
print(a.T)
print(a.T.shape)
print(a.ndim)
b=np.transpose(a)
print(b)
print(b.shape)
print(b.ndim)
```

Reshaping a array

```python
x=np.random.randint(10,50,(6,7))
print(x)
print(x.shape)
y=x.reshape(7,6)
print(y)
print(y.shape)
```

Mathematical operation with numpy arrays

```python
a=np.random.randint(20,50,(3,4))
b=np.random.randint(10,100,(3,4))

print(a)
print(a+5)
print(b)

#addition
print(a+b)
#subtraction
print(a-b)
#division
print(a/b)
#multiplication
print(a*b)
```

stacking of array

```python
a=np.arange(1,10)
b=np.arange(11,20)
print(a)
print(b)
```
